```python
from selenium import webdriver 
from selenium.webdriver.support.ui import Select
import time
import sys
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
import pandas as pd
```

```python
exeDriver = "chromedriver.exe"
driver = webdriver.Chrome(exeDriver)
print("Iniciando Scraper")
```

```python
url = "http://agrega.educacion.es/buscador2/FormularioAvanzadoCU/FormularioAvanzadoCU.do"
driver.get(url)
```

### Llenando campos de busqueda avanzada

```python
taxonomia = Select(driver.find_element_by_id('Taxonomia01')) 
taxonomia.select_by_value('NEDLOMES001') 

#nivelAgregacion = Select(driver.find_element_by_name('nivelAgreg'))
#nivelAgregacion.select_by_value('2')

tipoFormato = Select(driver.find_element_by_id('select')) 
tipoFormato.select_by_value('text/*') 

buscar = driver.find_elements_by_xpath("/html/body/div[1]/div/div/section/article[1]/section/form/fieldset/input[1]")[0]
buscar.click()
```

### Resultados de búsqueda

```python
columns = ['titulo', 'descripcion', 'formato', 'idioma', 'licencias', 'palabrasClave', 'url']
df = pd.DataFrame(columns=columns)

elementos = driver.find_elements_by_class_name("bloque_resultados") 
for elem in elementos:
    actions = ActionChains(driver)
    enlace = elem.find_elements_by_class_name("enlace")
    if len(enlace) != 0:
        titulo = enlace[0].text
        print(enlace[0].text)
        actions.key_down(Keys.CONTROL).click(enlace[0]).key_up(Keys.CONTROL).perform()
        driver.switch_to.window(driver.window_handles[-1])

        descripcion = driver.find_elements_by_xpath('/html/body/div[1]/div[2]/div/section/article/section[1]/p')[0]
        descripcion = descripcion.text

        formato = "texto"

        idioma = driver.find_elements_by_xpath('/html/body/div[1]/div[2]/div/section/article/section[3]/ul/li[2]/p[1]/span')[0]
        idioma = idioma.text

        licencias = driver.find_elements_by_xpath('/html/body/div[1]/div[2]/div/section/article/section[3]/ul/li[3]/p[1]/span')[0]
        licencias = licencias.text

        palabrasClave = driver.find_elements_by_xpath('/html/body/div[1]/div[2]/div/section/article/section[4]/ul/li/p/span')[0]
        palabrasClave = palabrasClave.text

        url = driver.current_url
    
        diccionario = {'titulo': titulo, 'descripcion': descripcion, 'formato': formato, 'idioma': idioma, 
                   'licencias': licencias, 'palabrasClave': palabrasClave, 'url': url}
        df = df.append(diccionario, ignore_index=True)
        
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        
```
